```python
#Scikit GMM trained on cleaned dataset. After testing various cluster sizes, 7 seems to produce the best results
import numpy as np
import matplotlib.pyplot as plt
from sklearn import mixture
from PIL import Image
import os
import os.path
import glob
%matplotlib inline

imgpathlist = glob.glob("all_vsi/*.jpg")
for img in imgpathlist:
    image = Image.open(img)
    image_data = np.array(image)
    h,w,c = image_data.shape
    data = image_data.reshape(h*w, c)
    gmm = mixture.GaussianMixture(n_components=7, covariance_type="tied")
    gmm = gmm.fit(data)
```

```python
#Example 1
image = Image.open("all_vsi/img_20200611_102539_045.jpg")
image_data = np.array(image)
h,w,c = image_data.shape
data = image_data.reshape(h*w, c)
cluster = gmm.predict(data)
cluster = cluster.reshape(h, w)
plt.imshow(cluster)
plt.title("img_20200611_102539_045.jpg")
plt.show()
```
